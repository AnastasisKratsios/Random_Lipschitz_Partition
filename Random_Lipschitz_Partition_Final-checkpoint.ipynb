# Random Lipschitz Partition

Given a positive integer $N$ and a real-number, this codes partitions the space into N random subsets such that:

 - Minimum probability that two nearby data-points are in different partitions,
 - Each has the same number of datapoints on average.
 
Can be used in Architopes (see: https://arxiv.org/abs/2006.14378) to make the regression algorithm semi-supervised.  


---------------------------------
Dataset used is the California Housing Market found here:
https://github.com/bzamanlooy/Architopes/tree/master/data

### Hyperparameters

```python
#Proportion of Total Data-Set Belonging to Each Partition on Average
N_partitions_to_make = 6

# maximum Iteration when building partition (once this number is hit we accept the partition size)
max_iter_per_parition_forming = 10**2

# Training/Testing Split Proportions
Train_to_Test_proportions = .7  #Note: Train/Test
```

## Initializations & Hyper-Parameters

### Import

```python
# Overall Use
%matplotlib inline
import numpy as np
import pandas as pd

# For Initialization of Initial Radius
from scipy.spatial.distance import pdist, squareform

# For Plotting
import matplotlib.pyplot as plt

# For Signaling User With Noise(s)
import os

# For Random Colouring
import random

# path expansion and existance checking
from pathlib import Path

# Training Test Split
from sklearn.model_selection import train_test_split
```

## Initialization(s) of Internal Variable(s), Dataset(s), and Function(s)

#### Seed

```python
# Set Seeds
random.seed(1001)
np.random.seed(1001)
```

#### Helper Function(s)

```python
# Function for Testing Random Partition
def is_in_random_ball(x_test,center,radius_in):
    TF_val = bool (np.linalg.norm((x_test-center), ord=1)<radius_in)
    return TF_val

# Delta (Radius...Deterministic Hyperparameter Part...)
def init_delta(X_train):
    X_train_subsample = X_train.loc[np.random.choice(X_train.index, 100, replace=False)]
    dis_sample = pdist(X_train_subsample[['longitude','latitude']] , 'minkowski', p=1)
    radius_delta_init = np.mean(dis_sample)/2
    
    return radius_delta_init

def check_path(path):
    path = os.path.expanduser(path)
    if not os.path.exists(path):
        Path(path).mkdir(parents=True, exist_ok=True)
    return(path)
```

#### Paths

```python
# Data Path (Read)
path_read = os.path.join('~', 'Desktop' , 'data_ocean_dist', 'housing.csv')
#path_read = check_path(path_read)

# Data Path (Write)salam khub
path_write = os.path.join('~', 'Dropbox', 'branching', 'data', 'random_partitioning/')
path_write = check_path(path_write)

### Datasets Reading and Writing
```

#### Pre-Process:
- Convert Categorical Variables to Dummies
- Remove Bad Column
- Perform Training/Test Split

```python
# Read Dataset & Drop Bad Column
X = pd.read_csv(path_read)#.drop(['total_bedrooms'], axis = 1)
X = pd.get_dummies(X)

X_train, X_test = train_test_split(X, test_size = (1-Train_to_Test_proportions), random_state = 2000)
X_train = X_train.reset_index().drop('index', axis = 1)
X_test = X_test.reset_index().drop('index', axis = 1)

# Create Categorical Columns
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
```

### Initialize Internal Parameters
Initialize Diameter Delta Parameter by subsampling from traning matrix, computing maximum distance between points, and dividing this quantity by 2.

```python
# Proportion of Paritions
Proportion_data_per_partition = (1/N_partitions_to_make)

# The amount of data which should be in each partition (approximately)...
N_training_samples = X_train.shape[0]
Data_size_test = N_training_samples*Proportion_data_per_partition

# Initialize Diameter of Random Balls
## Recall Longitude and Latitude Dimensions (For Plotting)
geo_facts = X[['longitude','latitude']]
geo_facts_maxs = np.max(geo_facts)
geo_facts_mins = np.min(geo_facts)
## Initialize Diameter
Diameter_Delta = init_delta(X_train)
```

### Build Random Partition

Steps:
1. Randomly Shuffle Training Data
2. Generate Random Radius
3. Build Balls Iteratively (Random Ball- Previous Ball: only if #Data is consistent across balls)

```python
# 1) Randomly Shuffle Data
X_train = X_train.sample(frac=1)

# 2) # Randomly Shuffle Data
rand_radius_factor = np.random.uniform(low = .25 , high = .5, size= 1)
random_radius = rand_radius_factor*Diameter_Delta
```

```python
# ---------------------------------------------------------------#
# Initialize Plots
#---------------------------------------------------------------#
# Create data
x = X[['longitude']]
y = X[['latitude']]

# plot Full Dataset
s_plot_area = .3
fig = plt.figure(num=None, figsize=(12, 12), dpi=80, facecolor='w', edgecolor='k')
ax1 = fig.add_subplot(111)
ax1.scatter(x, y, s=s_plot_area, c='black', alpha=0.1)

# Free Up Memory
#del X
#---------------------------------------------------------------#
# Begin Random Partitioning
#---------------------------------------------------------------#
# Announce Begining of Random Partitioning
# Print Spacers

print("----------------------------------------")
print("========================================")
# Report Number of Partitions and size of each
print("Total Data-Size: {}".format(N_training_samples))
print("Target Proportion of Data/Partition {}:".format(Proportion_data_per_partition))
# Print Spacers
print("========================================")
print("----------------------------------------")
print(" ")



# Initialize Current Training Data-Set Size
Current_Training_Set_Size = X_train.shape[0]
# Initialize Count Number of Partitions
n_partitions_realized = 0






while((Current_Training_Set_Size > (Data_size_test*1.5)) or 
      (n_partitions_realized<=(1/n_partitions_realized))): # Note we allow for some error in sampling so we don't have a tiny left-over dataset


    #---------------------------------------------------------------------------#
    # Build Current Partition: Initial Sampling Step
    #---------------------------------------------------------------------------#
    # Select Random Center
    center = X_train[['longitude','latitude']] # Only Use Geometrically-Meaningfull Coordinates!
    center = center.iloc[0]
    # Initialize Random Partition
    partition_train_indices = [is_in_random_ball(row, center, random_radius) for row in (X_train[['longitude','latitude']]).values]

    # Ensure that random partition is properly sized
    # Test Value
    test_partition_size = (np.sum(partition_train_indices)/ Data_size_test)
    # Failsafe
    number_iter = 0

    #---------------------------------------------------------------------------#
    # Build Current Partition: 'Ensure Correct Size' i.e.: Resampling Step
    #---------------------------------------------------------------------------#
    # Test if Partition is either too big or too small
    while(not (0.8 <= test_partition_size <= 1.2) and number_iter<max_iter_per_parition_forming):

        #print(test_partition_size)

        # Test if Should Resample Delta
        if(1.2 <= test_partition_size ): # Test if too Big
            #print("too big")
            # Resample Delta 
            Diameter_Delta = np.random.uniform(low = (Diameter_Delta/2) , high = Diameter_Delta, size= 1)
            random_radius = rand_radius_factor*Diameter_Delta
            # Rebuild Random Partition
            partition_train_indices = [is_in_random_ball(row, center, random_radius) for row in (X_train[['longitude','latitude']]).values]

        elif(test_partition_size <= .8): # Test if not too small
            #print("too small")
            # Resample Delta
            Diameter_Delta = np.random.uniform(low = Diameter_Delta , high = (2*Diameter_Delta), size= 1)
            random_radius = rand_radius_factor*Diameter_Delta
            # Rebuild Random Partition
            partition_train_indices = [is_in_random_ball(row, center, random_radius) for row in (X_train[['longitude','latitude']]).values]

        # Update Partition Test Size
        test_partition_size = (np.sum(partition_train_indices)/ Data_size_test)

        # Update Current Iteration_Count
        number_iter = number_iter + 1
        #print(test_partition_size)


    #---------------------------------------------------------------------------#
    ## Build Random Partition
    #---------------------------------------------------------------------------#

    ### Training Set
    partition_train = X_train.loc[partition_train_indices]
    partition_train = partition_train.reset_index().drop(['index'], axis = 1)
    #*
    Q_Train_or_Test = pd.DataFrame(np.repeat(True,(partition_train.shape[0])),columns={'is_train'})
    partition_train = pd.concat([partition_train.reset_index(),Q_Train_or_Test],axis=1)
    
    ### Build Test Set Partition
    partition_test_indices = [is_in_random_ball(row, center,random_radius) for row in (X_test[['longitude','latitude']]).values]
    partition_test = X_test.loc[partition_test_indices]
    partition_test = partition_test.reset_index().drop(['index'], axis = 1)
    #*
    Q_Train_or_Test = pd.DataFrame(np.repeat(False,(partition_test.shape[0])),columns={'is_train'})
    partition_test = pd.concat([partition_test.reset_index(),Q_Train_or_Test],axis=1)
    
    ## Creat Dataframe
    current_partition = pd.concat([partition_train,partition_test], axis = 0 )

    # Write Partition Dataframe to CSV
    path_write_partition = path_write + 'partition_' + str(n_partitions_realized) + '.csv'
    current_partition.to_csv(path_write_partition, index = False)
        
    
    
    #---------------------------------------------------------------------------#
    # Reports
    #---------------------------------------------------------------------------#
    
    # Report Current Partition Number 
    print("[ {} ]".format(n_partitions_realized))
    # Report New Partition Size
    print("New Training-Set Partition Size {} / {}".format((test_partition_size*Data_size_test),N_training_samples))
    # Report New Partition Proportion
    print("New Training-Set Partition Proportion {}".format(
        (round((test_partition_size*Proportion_data_per_partition),3))
        ))
    # Report Number of Iterations Needed to Stabilize New Partition within Perscribed Size
    print("Number of Sub-Iterations Needed {}".format(number_iter))
    # Report Mean Median Housing Price Stats within Partition
    print("Median Housing Price Stats Within Partition.  Mean:{}, St-Dev:{}, Range:{}".format(
        np.mean(partition_train[['median_house_value']]).values,
        np.std(partition_train[['median_house_value']]).values,
        (np.max(partition_train[['median_house_value']]).values - np.min(partition_train[['median_house_value']]).values)
        ))    
    # Print Spacer
    print("----------------------------------------")
    
    
    
    #---------------------------------------------------------------------------#
    ## Update Plot
    #---------------------------------------------------------------------------#
    # Initialize Current Colour-Vector
    partition_color = "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
    
    # Plot Training Data For New Cluster
    x = partition_train[['longitude']]
    y = partition_train[['latitude']]
    current_partition_train = 'K' + str(n_partitions_realized)
    ax1.scatter(x, y, s=s_plot_area, c=partition_color, alpha=0.5,label = str(current_partition_train))
    
    # Plot Test Data For New Cluster
    x = partition_test[['longitude']]
    y = partition_test[['latitude']]
    current_partition_test = 'tK' + str(n_partitions_realized)
    ax1.scatter(x, y, s=s_plot_area, c=partition_color, alpha=0.2, label =str(current_partition_test))
    
    
    
    #---------------------------------------------------------------------------#
    ## Remove Datapoint Belonging to Current Partition from Total Dataset
    #---------------------------------------------------------------------------#

    ### Filter Training Set
    downsample_indices = np.logical_not(partition_train_indices)
    X_train = X_train.loc[downsample_indices]

    ### Filter Test Set
    downsample_indices = np.logical_not(partition_test_indices)
    X_test = X_test.loc[downsample_indices]
    
    
    ## Update Current Training Data-Set Size
    Current_Training_Set_Size = X_train.shape[0]
    
    # Update Partition Count
    n_partitions_realized = n_partitions_realized + 1
    
    
    
    
    
    
    
    
#---------------------------------------------------------------------------#
# Process Remainder Points: i.e.: R^d - \cup_{i=1}^n K_i 
#---------------------------------------------------------------------------#
# Write Final Partition Separately
## Training Set
partition_train = X_train
partition_train = partition_train.reset_index().drop(['index'], axis = 1)
#*
Q_Train_or_Test = pd.DataFrame(np.repeat(True,(partition_train.shape[0])),columns={'is_train'})
partition_train = pd.concat([partition_train.reset_index(),Q_Train_or_Test],axis=1)

### Build Test Set Partition
partition_test = X_test
partition_test = partition_test.reset_index().drop(['index'], axis = 1)
#*
Q_Train_or_Test = pd.DataFrame(np.repeat(False,(partition_test.shape[0])),columns={'is_train'})
partition_test = pd.concat([partition_test.reset_index(),Q_Train_or_Test],axis=1)

## Creat Dataframe
current_partition = pd.concat([partition_train,partition_test], axis = 0 )

# Write Partition Dataframe to CSV
path_write_partition = path_write + 'partition_' + str(n_partitions_realized) + '.csv'
current_partition.to_csv(path_write_partition, index = False)



#---------------------------------------------------------------------------#
## Update Plot
#---------------------------------------------------------------------------#
# Update Current Partition Label
partition_color = "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])

# Plot Training Data For New Cluster
x = partition_train[['longitude']]
y = partition_train[['latitude']]
current_partition_train = 'K' + str(n_partitions_realized)
ax1.scatter(x, y, s=s_plot_area, c=partition_color, alpha=0.5,label = str(current_partition_train))

# Plot Test Data For New Cluster
x = partition_test[['longitude']]
y = partition_test[['latitude']]
current_partition_test = 'tK' + str(n_partitions_realized)
ax1.scatter(x, y, s=s_plot_area, c=partition_color, alpha=0.2, label =str(current_partition_test))









#---------------------------------------------------------------------------#
# Generate Reports for Final Partition
#---------------------------------------------------------------------------#
# Report For Final Random Partition
# Report Current Partition Number 
print("[ {} ]".format(n_partitions_realized))
# Report Final Partition Size
print("New Training-Set Partition Size {} / {}".format((partition_train.shape[0]),N_training_samples))
# Resport Final Partition Proportion
print("New Training-Set Partition Proportion {}".format(
    (round(((partition_train.shape[0]/N_training_samples)),3))
    ))
# Report Mean Median Housing Price Stats within Partition
print("Median Housing Price Stats Within Partition.  Mean:{}, St-Dev:{}, Range:{}".format(
    np.mean(partition_train[['median_house_value']]).values,
    np.std(partition_train[['median_house_value']]).values,
    (np.max(partition_train[['median_house_value']]).values - np.min(partition_train[['median_house_value']]).values)
    ))  
# Report Number of Iterations Needed to Stabilize
print("Number of Sub-Iterations Needed 0 (NA)")
# Print Spacer
print("----------------------------------------")

# Report Total Number of Partitions Used

# Print Spacers
print(" ")
print("----------------------------------------")
print("========================================")
# Report Number of Partitions and size of each
print("Total Number of Partitions Created: {}".format(n_partitions_realized + 1))
# Print Spacers
print("========================================")
print("----------------------------------------")







#-----------------------------------------------#
# Signal Code Completion with pop sound
#-----------------------------------------------#
beep = lambda x: os.system("echo -n '\a';sleep 0.2;" )
beep(1)


#plt.legend(loc='upper left');
```
